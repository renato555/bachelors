```python
import models
import datasets
import tensorflow as tf
import tensorflow.keras as keras
from os import path
```

# Hyperparameters search

```python
def hyperparameters_search(backbone_name, dataset_name, learning_rates, dropout_rates, log_path, image_size=224):
    data_train, data_val, data_test, info = datasets.load_data(dataset_name)
    data_train, data_val, data_test = datasets.preprocess_data(data_train, data_val, data_test, image_size)

    N_CLASSES = info.features['label'].num_classes
    epochs = 5
    for learning_rate in learning_rates:
        for dropout_rate in dropout_rates:
            model = models.get_classification_model(N_CLASSES=N_CLASSES,
                                                    backbone_name=backbone_name,
                                                    head_dropout_rate=dropout_rate)
            model.compile(loss=keras.losses.SparseCategoricalCrossentropy(),
                          optimizer=keras.optimizers.Adam(learning_rate=learning_rate),
                          metrics=["accuracy"])

            
            p = path.join(log_path, backbone_name, dataset_name, "search_L{}D_{}".format(learning_rate, dropout_rate))
            callbacks = [keras.callbacks.TensorBoard(p)]
            
            model.fit(data_train, epochs=epochs, validation_data=data_val, callbacks=callbacks)
```

```python
backbone_names = models.get_supported_backbones()
print(backbone_names)
dataset_names = datasets.get_supported_datasets()
print(dataset_names)
```

```python
learning_rates = [1e-4, 3e-4, 1e-3, 3e-3, 1e-2, 3e-2]
dropout_rates = [0.3, 0.5, 0.75]
for backbone_name in backbone_names:
    for dataset_name in dataset_names:
        hyperparameters_search(backbone_name, dataset_name, learning_rates, dropout_rates, "logs")
```

```python
learning_rates = [0.001]
dropout_rates = [0.2, 0.3, 0.45, 0.55, 0.7]
backbone_names = models.get_supported_backbones()
dataset_names = ["oxford_iiit_pet", "oxford_flowers102", "uc_merced"]
for backbone_name in backbone_names:
    for dataset_name in dataset_names:
        img_size = models.get_image_size(backbone_name)
        hyperparameters_search(backbone_name, dataset_name, learning_rates, dropout_rates, "dropout_search", img_size)
```
